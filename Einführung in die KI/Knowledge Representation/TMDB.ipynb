```python
import rdflib

g = rdflib.Graph()
g.parse('data/TMDB.ttl', format='turtle')

query_a = g.query("""
SELECT distinct ?gg
WHERE {
    ?m a :Movie ;
    :genres ?g .
?g  :name ?gg.
}
""")
for row in query_a:
    print(row.gg)
```

```python
query_b = g.query("""
SELECT distinct ?dep
WHERE {
    ?m a :Movie ;
    :crew ?c .
?c  :department ?dep.
}
""")
for row in query_b:
    print(row.dep)
```

```python
query_c = g.query("""
SELECT (COUNT(DISTINCT ?num) AS ?count)
WHERE {
    ?m a :Movie ;
    :production_companies ?com .
?com  :id ?num.
}
""")
for row in query_c:
    print(row[0])
```

```python
query_d = g.query("""
SELECT distinct ?cl
WHERE {
    ?m a :Movie ;
    :belongs_to_collection ?col .
?col  :name ?cl.
}
""")
for row in query_d:
    print(row.cl)
```

```python
query_e = g.query("""
SELECT ?t
WHERE {
    ?m a :Movie ;
    :original_language "de";
    :title ?t.
}
""")
for row in query_e:
    print(row.t)
```

```python
query_f = g.query("""
SELECT ?t
WHERE {
    ?m :cast ?c .
    ?c :character "Julius Caesar" .
    ?m :title ?t .
}
""")
for row in query_f:
    print(row.t)
```

```python
query_g = g.query("""
SELECT DISTINCT ?p
WHERE {
    ?mov :cast ?actor .
    ?mov :crew ?director .
    ?actor :character ?p .
    ?director :name ?p .
    ?director :job "Director" .
}
""")
for row in query_g:
    print(row.p)
```

```python
query_h = g.query("""
SELECT DISTINCT ?role
WHERE {
    ?mov :cast ?actor .
    ?actor :character ?role .
    ?actor :name ?pers .
    FILTER(?pers = "Jason Statham")
}
""")
for row in query_h:
    print(row.role)
```

```python
query_i = g.query("""
SELECT DISTINCT ?t ?ot ?rd ?rt ?ov ?tl ?bg ?rv ?pp ?lg ?ctp ?cpp ?gn
WHERE {
    ?movie a :Movie ;
        :title ?t ;
        :original_title ?ot ;
        :release_date ?rd ;
        :runtime ?rt ;
        :overview ?ov ;
        :tagline ?tl ;
        :budget ?bg ;
        :revenue ?rv ;
        :popularity ?pp ;
        :original_language ?lg .
    OPTIONAL { ?movie :production_countries ?country .
        ?country :name ?ctp . }
    OPTIONAL { ?movie :production_companies ?company .
        ?company :name ?ccp . }
    OPTIONAL { ?movie :genres ?genre .
        ?genre :name ?gn . }
    FILTER(?t = "American Beauty")
}
""")

for row in query_i:
    print(f"{row.t} | {row.ot} | {row.rd} | {row.rt} | {row.ov} | {row.tl} | {row.bg} | {row.rv} | {row.pp} | {row.lg} | {row.ctp} | {row.cpp} | {row.gn}")
```
